```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras import Sequential
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
plt.matshow(df.corr())
plt.show()
```

```python
min_max_scaler = MinMaxScaler()
df_scaled = min_max_scaler.fit_transform(df.values)
df = pd.DataFrame(df_scaled)
```

```python
x_df = df.iloc[:,:-1]
```

```python
y_df = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_df,y_df,test_size=0.2,random_state=42)
```

```python
model = Sequential()
model.add(Dense(130,activation='selu',input_shape=[13]))
model.add(Dense(65,activation='selu'))
model.add(Dropout(0.2))
model.add(Dense(65,activation='selu'))
model.add(Dropout(0.2))
model.add(Dense(65,activation='selu'))
model.add(Dropout(0.2))
model.add(Dense(65,activation='selu'))
model.add(Dropout(0.2))
model.add(Dense(13,activation='selu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=1000)
```

```python
acc = history.history['accuracy']
loss = history.history['loss']
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']

epoch = range(len(acc))

plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
plt.plot(epoch,acc,label='acc')
plt.plot(epoch,val_acc,label='val_acc')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epoch,loss,label='loss')
plt.plot(epoch,val_loss,label='val_loss')
plt.title('Loss')
```
